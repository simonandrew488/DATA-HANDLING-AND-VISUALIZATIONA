```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[4,5,6,7]
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line graph")
plt.plot(x,y)
plt.show()
```

```python
x=[20,30,40,50]
y=["bananas","apples","dates","cherries"]
c=["black","pink","yellow","blue"]
d=[0,0.2,0,0.1]
plt.pie(x,labels=y,colors=c,explode=d,shadow=2)

plt.show()
```

```python
import numpy as np
x=np.array(["a","b","c","d"])
y=np.array([3,2.5,1,6])
z=["red","purple","green","orange"]
plt.bar(x,y,color=z)
plt.show
```

```python
x=np.array([1,4,7,3])
y=np.array([3,2,1,6])
z=["red","purple","green","orange"]
plt.scatter(x,y,c=z)
plt.show()
```

```python
x=np.array([1,4,7,3])
y=np.array([3,2,1,6])
z=["red","purple","green","orange"]
a=[30,40,50,60]
plt.scatter(x,y,a,c=z,edgecolor="r",marker="*")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,5,6,7]
colors=["r","y","b","g","b"]
sizes=[100,200,300,40,15]
plt.scatter(x,y,c=colors,s=sizes,alpha=0.9,marker="*",edgecolor="green")
plt.title("scatter",fontsize=20)
plt.xlabel("x axis",fontsize=20)
plt.ylabel("y axis",fontsize=20)

```
